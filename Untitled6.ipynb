```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import math
from scipy.stats import norm

class Task:
    def __init__(self, name, optimistic, most_likely, pessimistic):
        self.name = name
        self.optimistic = optimistic
        self.most_likely = most_likely
        self.pessimistic = pessimistic
        self.expected_time = (optimistic + 4 * most_likely + pessimistic) / 6
        self.variance = ((pessimistic - optimistic) / 6) ** 2

def create_pert_graph(tasks, dependencies):
    G = nx.DiGraph()

    for task in tasks:
        G.add_node(task.name, expected_time=task.expected_time, variance=task.variance)

    for dep in dependencies:
        for to_task in dep[1]:
            G.add_edge(dep[0], to_task, weight=G.nodes[dep[0]]['expected_time'])

    return G

def find_critical_path(G):
    critical_path = nx.dag_longest_path(G, weight='weight')
    critical_path_length = sum(G.nodes[node]['expected_time'] for node in critical_path)
    return critical_path, critical_path_length

def calculate_guaranteed_interval(critical_path_length, G, P):
    sum_variances = sum(G.nodes[node]['variance'] for node in critical_path)
    standard_deviation_critical_path = math.sqrt(sum_variances)

    # Находим z для заданной вероятности P
    z_value = norm.ppf(P)

    # Вычисляем интервал гарантированного времени выполнения проекта
    guaranteed_interval = z_value * standard_deviation_critical_path

    return critical_path_length - guaranteed_interval, critical_path_length + guaranteed_interval

def draw_pert_graph(G, critical_path):
    plt.figure(figsize=(14, 10))  # Увеличиваем размер фигуры

    pos = graphviz_layout(G, prog='dot')
    labels = {node: f"{node}\n{G.nodes[node]['expected_time']:.2f}" for node in G.nodes}
    edge_labels = {(u, v): f"{G[u][v]['weight']:.2f}" for u, v in G.edges}

    nx.draw(G, pos, with_labels=True, labels=labels, node_color='lightblue', node_size=2000, font_size=10)
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    nx.draw_networkx_edges(G, pos, edgelist=list(zip(critical_path[:-1], critical_path[1:])), edge_color='r', width=2)

    # Добавляем информацию о критическом пути и максимально возможном сроке выполнения проекта
    plt.text(0.5, -0.1, f'Critical Path: {critical_path}\nCritical Path Length: {critical_path_length:.2f} days',
             horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.text(0.5, -0.2, f'Max Possible Time (99% reliability): {max_possible_time:.2f} days',
             horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

    plt.show()

# Пример данных
tasks = [
    Task('Start', 0, 0, 0),
    Task('b1', 3, 6, 10),
    Task('b2', 2, 4, 7),
    Task('b3', 2, 5, 8),
    Task('b4', 1, 3, 6),
    Task('b5', 2, 4, 7),
    Task('b6', 3, 8, 10),
    Task('b7', 5, 10, 15),
    Task('b8', 3, 6, 9),
    Task('b9', 2, 4, 7),
    Task('b10', 5, 9, 11),
    Task('b11', 2, 7, 9),
    Task('End', 0, 0, 0)
]

dependencies = [
    ('Start', ['b1']),
    ('Start', ['b2']),
    ('b1', ['b3']),
    ('b2', ['b4', 'b7']),
    ('b3', ['b5', 'b6']),
    ('b4', ['b5', 'b6']),
    ('b5', ['b8', 'b9', 'b10']),
    ('b6', ['b9', 'b10']),
    ('b7', ['b9', 'b10']),
    ('b8', ['b11']),
    ('b9', ['b11']),
    ('b11', ['End'])
]

# Создание графа PERT
G = create_pert_graph(tasks, dependencies)

# Нахождение критического пути
critical_path, critical_path_length = find_critical_path(G)

# Рассчитываем максимально возможный срок выполнения проекта с надежностью 0.99
gamma = 0.99
min_possible_time, max_possible_time = calculate_guaranteed_interval(critical_path_length, G, gamma)
P = 0.9973
min_time, max_time = calculate_guaranteed_interval(critical_path_length, G, P)

# Отображение диаграммы PERT с информацией о критическом пути и максимально возможном сроке выполнения проекта
print(f'Critical Path: {critical_path}')
print(f'Critical Path Length: {critical_path_length:.2f} days')
print(f'Max Possible Time (99% reliability): {max_possible_time:.2f} days')
print(f'Guaranteed Time Interval (P = 0.9973): [{min_time:.2f}, {max_time:.2f}] days')
draw_pert_graph(G, critical_path)
```
